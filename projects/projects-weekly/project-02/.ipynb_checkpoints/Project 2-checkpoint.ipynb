The dataset that I worked with for this project represented those billboard tracks which peaked in 2000. There were some tracks from 1999 and some from 2000, but all tracks only reached their best rankings on the charts in 2000. The data was poorly entered and poorly organized, so some changes had to be made. I began by changing the columns to shorter, more appropriate names. I then converted data types for datetime objects. There were multiple mispellings of genres, so I replaced the mispellings with the proper spellings. There were also mispellings for artist names, so I replaced those as well. I noticed that some songs had been incorrectly categorized to a genre, but I decided to not change this data, as I could not be sure of where this information was obtained. It could be the case that the data was incorrect, but it could also be the case that those genre labels came from a reputable source. Lastly, I replaced * with NaN.

As for the new columns I added, I created columns pertaining to the date the track peaked, the month the track peaked, the peak rank, the number of weeks the track was at its peak rank, and the total number of weeks the track was in the chart. Lastly, I created a new data frame the counted the number of tracks peaking by month, sorted by genre.

I grouped together all of the functions and manipulations I used to get the data to where it needed to be in order to be useful for analysis. I also included most of the manipulations used for the graphs, in order to be succinct.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime 
%matplotlib inline

def t_sec(x):
    x=str(x)
    m=x[0]
    s=x[1:3]
    m=int(m)
    s=int(s)
    t=m*60+s
    return t

df=pd.read_csv("/Users/thomas/GA-DSI/projects/projects-weekly/project-02/assets/billboard.csv")
name=range(1,77)
names=["week_"+str(x) for x in name]

#Renaming columns; made easier by using range function and concatenation
df.columns=['year','artist','track_name','track_time','genre','date_entered','date_peaked','week_1', 'week_2', \
            'week_3', 'week_4', 'week_5', 'week_6', 'week_7', 'week_8', 'week_9', 'week_10', 'week_11', 'week_12',\
            'week_13', 'week_14', 'week_15', 'week_16', 'week_17', 'week_18', 'week_19', 'week_20', 'week_21', \
            'week_22', 'week_23', 'week_24', 'week_25', 'week_26', 'week_27', 'week_28', 'week_29', 'week_30', \
            'week_31', 'week_32', 'week_33', 'week_34', 'week_35', 'week_36', 'week_37', 'week_38', 'week_39', \
            'week_40', 'week_41', 'week_42', 'week_43', 'week_44', 'week_45', 'week_46', 'week_47', 'week_48', \
            'week_49', 'week_50', 'week_51', 'week_52', 'week_53', 'week_54', 'week_55', 'week_56', 'week_57', \
            'week_58', 'week_59', 'week_60', 'week_61', 'week_62', 'week_63', 'week_64', 'week_65', 'week_66', \
            'week_67', 'week_68', 'week_69', 'week_70', 'week_71', 'week_72', 'week_73', 'week_74', 'week_75', \
            'week_76']

#Change time to seconds
df["track_time"]=df["track_time"].apply(lambda x: t_sec(int(x[0:7].replace(',',''))))

#Change Genre Names
df['genre'].replace('R & B','R&B', inplace=True)
df['genre'].replace("Rock'n'roll","Rock",inplace=True)

#Datetime
df['date_entered']=pd.to_datetime(df['date_entered'])
df['date_peaked']=pd.to_datetime(df['date_peaked'])

#Artist Names
df["artist"].replace("Janet","Janet Jackson",inplace=True)
df["artist"].replace("2Ge+her","2Gether",inplace=True)
df["artist"].replace("matchbox twenty","Matchbox Twenty",inplace=True)

#Take * and replace with NaN
df.replace('*',np.nan,inplace=True)
#New column for days between date_entered and date_peaked
df["days_until_peak"]= [int(datetime.strftime(x,"%j"))-int(datetime.strftime(y,"%j")) if  \
       int(datetime.strftime(x,"%j"))-int(datetime.strftime(y,"%j"))>=0 else \
       int(datetime.strftime(x,"%j"))-int(datetime.strftime(y,"%j"))+365 \
                    for x,y in zip(df['date_peaked']\
                                   , df['date_entered'])]

#Rearranging columns and finding lowest rank acheived
cols=df.columns.tolist()
cols.insert(7,cols[-1:][0])
cols=cols[:-1]
df=df[cols]

cols=df.columns.tolist()
cols=cols[8:]

low_list=[]
i=0

while i<317:
    low=[]
    for col in cols:
        low.append(df.loc[i,col])
    low_list.append(low)
    i+=1

i=0
j=0

while j<len(low_list):
    i=0
    while i<len(low_list[j]):
        if type(low_list[j][i])==str:
            low_list[j][i]=int(low_list[j][i])
        i+=1
    j+=1
#Finds lowest rank achieved
lowest_rank = [min (x) for x in low_list]
#Counts number of time lowest rank occurs
lowest_rank_count = [x.count(y) for x,y in zip(low_list,lowest_rank)]
#Counts total weeks in rankings
total_weeks=[]
for x in low_list:
    i=0
    for y in x:
        if np.isnan(y):
            pass
        else:
            i+=1

    total_weeks.append(i)

#Put new lists in to df
temp_df1=pd.DataFrame(lowest_rank,columns=['lowest_rank'])
temp_df2=pd.DataFrame(lowest_rank_count,columns=['lowest_rank_count'])
temp_df3=pd.DataFrame(total_weeks,columns=['total_weeks'])
df=pd.concat([df,temp_df1,temp_df2,temp_df3],axis=1)
cols=df.columns.tolist()
cols.insert(8,cols[-3:][0])
cols.insert(9,cols[-2:][0])
cols.insert(10,cols[-1:][0])
cols=cols[:-3]
df=df[cols]

#The month of the date peaked
df['date_peaked']=pd.to_datetime(df['date_peaked'])
month_peaked=[x.month for x in df['date_peaked']]
temp_df=pd.DataFrame(month_peaked,columns=['month_peaked'])
df=pd.concat([df,temp_df],axis=1)
cols=df.columns.tolist()
cols.insert(7,cols[-1:][0])
cols=cols[:-1]
df=df[cols]

#This will be used for the stacked bar chart
df_temp=pd.melt(df,id_vars="month_peaked",value_vars=['genre'])
df_temp
cols=df['genre'].unique()
l=[]
for x in cols:
    l.append([x[0] for x in df_temp[df_temp['value']==x].values.tolist()])
g=[]
c=0
h={}
for i in l:
    g=[]
    for x in range(1,13):
        count=i.count(x)
        a= x
        b= count
        g.append(b)
    h[cols[c]]=g
    c+=1
        
dfh=pd.DataFrame(h)
```

I began with some exploratory analysis to see which genres were most popular. I didn't feel that it was a good idea to go on a track or artist basis due to the number of unique tracks and unique artists. I felt it was much better to go with the genre basis due to the low number of unique genres.

The first barchart is the average peak ranking by genre. Jazz seems to be extraordinarily successful, whereas R&B seems to be somewhat a poor choice for #1 rankings.

The second barchart demonstrates that there are next to no Jazz tracks, so the Jazz average ranking is clearly biased. Furthermore, Rock, Country, Rap, and R&B are the most frequent genres of music on the top 100, indicating that they are popular and have a high rate of success. One cannot be certain of how many tracks not on the top 100 are made by genre, so it would not be fair to firmly say that a certain kind of music is more successful than another, but it would be fair to say that a certain kind of music is more popular than another.

```python
genres=['Rock', 'Latin', 'Country', 'Rap', 'Pop', 'Electronica', 'Jazz', 'R&B', 'Reggae', 'Gospel']
avg_ranks=[(x,df[df['genre']==x]['lowest_rank'].mean(),df[df['genre']==x]['lowest_rank'].count()) for x in genres]
avg_ranks.sort(key=lambda x: x[1])
genres=[x[0] for x in avg_ranks]
average=[x[1] for x in avg_ranks]
tracks=[x[2] for x in avg_ranks]
plt.subplot(211)
plt.bar(range(10),average)
plt.xticks(np.linspace(0.2,9.2,10),genres,rotation=45)
plt.ylabel("Average Peak Ranking")
plt.title("Average Peak Ranking by Genre")
plt.show()
plt.subplot(212)
plt.bar(range(10),tracks)
plt.xticks(np.linspace(0.2,9.2,10),genres,rotation=45)
plt.title("Number of Tracks in Charts by Genre")
plt.ylabel("Number of Tracks")
plt.show()
```

My first plot was a barchart of the weeks that a track was at its lowest rank. (NOTE: The barplot skips 6:9 for the convenence of plotting). The plot indicates that a track stays at its peak for a very brief period, with most tracks remaining at their peak for only one week, and nearly all tracks remaining at their peak for three weeks or less. Only two tracks reached 10 and 11 weeks, so they are outliers. The observed shape is highly representative of a poisson distribution.

The next plot is a scatter plot of the relation between the day until the track peaks and the total number of weeks that the track is in the rankings. This graph is somewhat trivial, as one would expect the number of days a track takes to peak to directly influence the total number of days. However, this graph is shown to further build upon the correlation found in the previous chart, that tracks stay at their peak for a minimal time. This graph demonstrates how tracks leaving their peak do not stay for extended periods of time on the top 100 rankings, but rather fall off quite quickly and consistently.

The third chart, titled 'Days until Peak vs Lowest Rank', demonstrates another important fact; the lower the peak rank, the more days it takes to peak. So, tracks that are rank #1 material will need to be released with extra time to compensate for the descent time it takes to reach rank #1.

The fourth chart, titled 'Total Weeks in Chart by Lowest Rank', further details the total weeks and lowest rank relation, except it reveals the genres of each data point. This chart is highly significant as it reveals that songs of the 'Rock' genre are highly successful. It was already pointed out that Rock tracks have the lowest average peak ranking, but this chart demonstrates just how long 'Rock' tracks are on the top 100. There is a large cluster of data points at around week 20, so some more analysis will need to be done to see just how many tracks only last 20 weeks.

In the fifth plot, which is a histogram, I've plotted the number of weeks a track has been in the charts; the y-axis simply denotes the number of tracks which fall into each bin. As one can see, a considerable number of tracks last exactly 20 weeks in the charts. 

In the sixth and seventh plots, I further analyze the tracks at week 20, since they make up at least 1/4 of all tracks. The sixth plot plots the days until a track peaks, and the seventh plot plots the lowest rank achieved. Plot six looks fairly Gaussian with a mean around 70 days, and plot seven looks more like a Poisson distribution or skewed Gaussian. In short, one can expect these tracks to follow on a recurring distribution.

```python
#T-Test

i=0
j=0
nums=[]
num=[]
while i in range (12):
    for x in dfh:
        num.append(dfh.loc[i,x])
    numsum=sum(num)
    nums.append(numsum)
    num=[]
    numsum=[]
    i+=1
nums
#Do a quick T-Test [µhat-µ]/[s/sqrt[n]]
uhat=(nums[0]+float(nums[-1]))/2
u=np.mean(nums)
n=len(nums)
s=np.std([nums[0],nums[-1]])

t_test=(uhat-u)/(s/np.sqrt(n))
print "T Test Value is: ",t_test
print "Degrees of Freedom:",n-1
print "For Alpha of 0.05, t-test value is: 1.796"
print "Test Rejects Null Hppothesis"
```
