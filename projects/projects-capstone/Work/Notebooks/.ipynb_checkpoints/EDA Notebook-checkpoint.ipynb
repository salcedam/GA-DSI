```python
df=pd.read_json('../r_news.json')
for i in range(len(df.columns)):
    comments=df.iloc[4,i]
    all_words=""
    for x in comments:
        if len(x)>1:
            for y in x:
                all_words+=y['comment']+" "
        else:
            if type(x[0])==dict:
                all_words+=x[0]['comment']+" "
            else:
                print x[0]
    df.iloc[4,i]=all_words
```
