```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import ExtraTreesClassifier,GradientBoostingClassifier,RandomForestClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import HashingVectorizer
from sklearn.linear_model import Ridge,LassoCV
from sklearn.cross_validation import train_test_split
from nltk.stem import SnowballStemmer,PorterStemmer
from stemming import lovins,porter2,paicehusk
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
import stop_words
import pprint
import re
import time
import spacy
import datetime
%matplotlib inline
```

```python
s_p=pd.read_csv('../s&p_index_prices.csv')
s_p["Date"]=s_p["Date"].apply(lambda x: datetime.datetime.strptime(x,'%Y-%m-%d'))
s_p.head(2)
```

```python
vol=np.mean(s_p["Volume"])/25
plt.plot_date(s_p["Date"],(s_p["High"]-s_p["Low"]),'-',label="Volatility Measure")
plt.plot_date(s_p["Date"],s_p["Volume"]/(vol),"-k",label=("Volume"))
plt.legend()
plt.show()
```

```python
plt.plot_date(s_p["Date"],s_p["Open"],'-')
plt.title("S&P Historic Prices")
plt.show()
```

```python
companies=pd.read_csv("../s&p_stock_info.csv")
companies.head(2)
```

```python
sns.barplot(companies['Sector'].value_counts().index,companies['Sector'].value_counts().values)
plt.xticks(rotation=90)
plt.title("Companies per Sector")
plt.show()
sns.barplot(companies.groupby("Sector")['Price'].mean().index,\
            companies.groupby("Sector")['Price'].mean().values)
plt.xticks(rotation=90)
plt.title("Average Price")
plt.show()
sns.barplot(companies.groupby("Sector")['Earnings/Share'].mean().index,\
            companies.groupby("Sector")['Earnings/Share'].mean().values)
plt.xticks(rotation=90)
plt.title("Average Earning/Share")
plt.show()
sns.barplot(companies.groupby("Sector")["Market Cap"].sum().index,\
        companies.groupby("Sector")["Market Cap"].sum().values)
plt.xticks(rotation=90)
plt.title("Market Caps ($B)")
plt.show()
```

```python
gold=pd.read_csv("../Historic_Gold.csv")
crude=pd.read_csv("../Historic_Crude.csv")
gold["Date"]=pd.to_datetime(gold["Date"])
gold["Date"].dtype
crude["DATE"]=pd.to_datetime(crude["DATE"])
crude["DATE"].dtype
crude=crude[crude["VALUE"]!="."]
crude["VALUE"]=crude["VALUE"].astype(float)
prices=pd.merge(crude,gold,left_on="DATE",right_on='Date')
prices=prices[pd.notnull(prices["Vol"])]
prices["Vol"]=prices["Vol"].apply(lambda x: 0. if x=='-' else x)
# prices=prices[pd.notnull(prices[prices["Vol"]!='-']["Vol"])]
prices["Vol"]=prices["Vol"].apply(lambda x: float(x[:-1]) if x!=0 else x)
prices=prices.merge(s_p,left_on="DATE",right_on='Date')
```

```python
prices
```

```python
plt.plot_date(crude["DATE"],crude["VALUE"]*10+1000,'-',label="Crude",c="black")
plt.plot_date(gold["Date"],gold["Open"]+400,'-',label="Gold",c="gold")
plt.plot_date(s_p["Date"],s_p["Open"]-850,'-',label="S&P",c='green')
plt.title("Historic Prices")
plt.legend(loc='lower right')
plt.yticks([0])
plt.show()
```

```python
x_train,x_test,y_train,y_test=train_test_split(prices[['VALUE','Price']],prices['Open_y'])
linreg=LassoCV()
linreg.fit(x_train,y_train)
prediction=linreg.predict(x_test)
print linreg.score(x_test,y_test)
plt.plot(range(len(prediction)),(prediction-y_test)/y_test,'-')
plt.show()
```
