```python
df=data.reset_index()
del df['Date']
y=df['Weekly_Sales']
X=df[['Store','Dept','IsHoliday']]
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(X,y)
```

```python
model=LinearRegression().fit(xtrain,ytrain)
pred=model.predict(xtest)
```

```python
plt.plot(ytest-pred,'o');
```

### Lets try doing a linear regression by department

```python
df=master.copy()
scores=[]
for department in master.Dept.unique():
    y=master[master['Dept']==department]['Weekly_Sales']
    X=master[master['Dept']==department][['IsHoliday', u'CPI', u'Unemployment',u'Temperature',
           u'Fuel_Price']]
    xtrain,xtest,ytrain,ytest=train_test_split(X,y)
    model=LinearRegression().fit(xtrain,ytrain)
    pred=model.predict(xtest)
    plt.plot(ytest-pred,'o')
    scores.append(model.score(xtest,ytest))
# plt.plot([0,200000],[0,200000],c='red',linestyle='-')
print np.mean(scores)
plt.show()
```

### By Department

```python
for department in data.Dept.unique():
    if len(data[data['Dept']==department])>10:
        df=data[data['Dept']==department].copy()
        df['Date2']=pd.to_datetime(df.index)
        df["Week"]=[x.week for x in df["Date2"]]
        dfg=df.groupby(["Week"])['Weekly_Sales'].mean()
        plot_acf(dfg)
        plt.xlabel(department)
plt.show()
```

### Findings

Some departments are seasonally periodic, whereas others are weekly periodic. Regard autocorrelation functions above. (Note that the functions were not plotted to save RAM)
