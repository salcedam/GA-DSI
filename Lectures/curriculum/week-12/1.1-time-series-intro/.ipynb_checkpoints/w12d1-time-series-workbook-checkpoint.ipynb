## Stock Prediction

```python
sp_start = datetime.datetime(2000, 01, 01)
sp_end = datetime.datetime(2016, 07, 31)
sp = web.DataReader('SPY', 'google', sp_start, sp_end)
```

```python
sp
```

```python
for i in range(1, 21, 1):
    sp.loc[:,'Close Minus ' + str(i)] = sp['Close'].shift(i)
    sp20 = sp[[x for x in sp.columns if 'Close Minus' in x or x == 'Close']].iloc[20:,]
    sp.loc[:,'Open Minus ' + str(i)] = sp['Open'].shift(i)
    sp40 = sp[[x for x in sp.columns if 'Open Minus' in x or x == 'Open']]
sp40['3-day_MA'] = np.mean(sp40.iloc[:,1:4].T)
sp40['5-day_MA'] = np.mean(sp40.iloc[:,1:6].T)
sp40['10-day_MA']= np.mean(sp40.iloc[:,1:11].T)
sp40['15-day_MA']= np.mean(sp40.iloc[:,1:16].T)
sp40['20-day_MA']= np.mean(sp40.iloc[:,1:21].T)

sp20=sp20.merge(sp40,left_index=True,right_index=True)
for col in sp20.columns[1:,]:
    sp20[col]=sp20['Close']-sp20[col]
sp20
```

```python
sp20
```

```python
sp20['Open']=sp['Open'].copy()
sp20['Day_Delta']=sp20['Close']-sp20['Open']
sp20=sp20.reset_index()
sp20=sp20.dropna()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import ExtraTreesRegressor,GradientBoostingRegressor
```

```python
columns=[u'Close Minus 1', u'Close Minus 2', u'Close Minus 3',u'Open Minus 1', u'Open Minus 2',
       u'Open Minus 3', u'5-day_MA']


y_train=sp20['Day_Delta'].iloc[:3500].copy()
y_test =sp20['Day_Delta'].iloc[3500:3900].copy()
x_train=sp20[columns].iloc[:3500,]
x_test =sp20[columns].iloc[3500:3900,]
```

```python
model=LinearRegression().fit(x_train,y_train)
pred=model.predict(x_test)
```

```python
plt.plot(range(len(y_test)),pred-y_test)
plt.show()
plt.plot(y_test,pred,'o')
plt.xlabel('Y-Test')
plt.ylabel('Predictions')
plt.show()
```

```python
profits=[]
for value1,value2 in zip(pred,y_test):
    if value1>0:
        profits.append((value2)-1)
sum(profits)
```

```python
model.score(x_test,y_test)
```

```python
model.coef_
```

```python
cdc = sp[['Close']].iloc[-1000:]
ndo = sp[['Open']].iloc[-1000:].shift(-1)
```

```python
tf1 = pd.merge(tf, cdc, left_index=True, right_index=True)
tf2 = pd.merge(tf1, ndo, left_index=True, right_index=True)

tf2.columns = ['Next Day Close', 'Predicted Next Close', 'Current Day Close', 'Next Day Open']
tf2
```

```python
def get_signal(r):
    if r['Predicted Next Close'] > r['Next Day Open']:
        return 1
    else:
        return 0

def get_ret(r):
    if r['Signal'] == 1:
        return (r['Next Day Close'] - r['Next Day Open'])
    else:
        return 0

tf2 = tf2.assign(Signal = tf2.apply(get_signal, axis=1))
tf2 = tf2.assign(PnL = tf2.apply(get_ret, axis=1))

tf2
```
