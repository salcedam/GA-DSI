```python
#Using 'with ...' eliminates the need to close the file; same functionality
#f.printline() will read the current line, and then move the placeholder of the file to the next line;
    #you can't go back
#for line in f: ... will read all of the remaining lines in f, but will not move the placeholder to 
    #the next line like printline() does 
data=""
o=open("/Users/thomas/GA-DSI/curriculum/week-01/3.1-notebooks-csvs/assets/datasets/sales_copy.txt",'r')
with open("/Users/thomas/GA-DSI/curriculum/week-01/3.1-notebooks-csvs/assets/datasets/sales_copy_copy.txt",'w+') as f:
    for line in o:
        f.write(line) #Note that writing with 'w' or 'w+' will wipe the initial file and rewrite
o.close()
```
