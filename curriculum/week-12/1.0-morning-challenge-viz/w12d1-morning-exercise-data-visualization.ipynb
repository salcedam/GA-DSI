## Doing some timeseries plots

```python
columns=['inbound_sms_count','inbound_vc_minutes','outbound_sms_count','outbound_vc_minutes',\
         'in_network_inbound_sms_count',\
'out_of_network_inbound_sms_count','in_network_outbound_sms_count','out_of_network_outbound_sms_count',\
'in_network_inbound_vc_minutes','out_of_network_inbound_vc_minutes','in_network_outbound_vc_minutes',\
'out_of_network_outbound_vc_minutes']
for col in columns:
    plt.plot_date(usage['timestamp'],usage[col])
    plt.ylabel(col)
    plt.show()
```

```python
X=usage[columns]
y=usage['timestamp']
```

```python
x0=X.copy()
x1=pd.DataFrame(preprocessing.RobustScaler().fit_transform(X),columns=columns)
x2=pd.DataFrame(preprocessing.MinMaxScaler().fit_transform(X),columns=columns)
x3=pd.DataFrame(preprocessing.Normalizer().fit_transform(X),columns=columns)
```

```python
# sns.pairplot(x1);
```

```python
for col in x1:
    plt.plot_date(y,x1[col])
    plt.plot_date(y,x2[col])
    plt.plot_date(y,x3[col])
    plt.show()
```

```python
usage['out_of_network_vc_total']=usage['out_of_network_inbound_vc_minutes']+usage['out_of_network_outbound_vc_minutes']
usage['in_network_vc_total']=usage['in_network_inbound_vc_minutes']+usage['in_network_outbound_vc_minutes']
usage['out_of_network_sms_total']=usage['out_of_network_inbound_sms_count']+usage['out_of_network_outbound_sms_count']
usage['in_network_sms_total']=usage['in_network_inbound_sms_count']+usage['in_network_outbound_sms_count']
```

```python
usage['day']=\
    usage['timestamp'].apply(lambda x:datetime.datetime.strftime(datetime.datetime.strptime(x,'%Y-%m-%d'),'%w'))
```

```python
i_1=usage.groupby('day')['inbound_sms_count','outbound_sms_count'].mean()['inbound_sms_count']
i_2=usage.groupby('day')['inbound_sms_count','outbound_sms_count'].mean()['outbound_sms_count']

plt.bar([x-.4 for x in [1,2,3,4,5,6,7]],i_1,width=.4,color='blue',label='Inbound sms')
plt.bar([1,2,3,4,5,6,7],i_2,width=.4,color='grey', label='Outbound sms')
plt.legend(loc='upper left')
plt.xticks(range(1,8),['SUN','MON','TUE','WED','THU','FRI','SAT'])
plt.show()

i_3=usage.groupby('day')['out_of_network_sms_total','in_network_sms_total'].mean()['out_of_network_sms_total']
i_4=usage.groupby('day')['out_of_network_sms_total','in_network_sms_total'].mean()['in_network_sms_total']
plt.bar([x-.4 for x in [1,2,3,4,5,6,7]],i_4,width=.4,color='blue', label='In network sms')
plt.bar([1,2,3,4,5,6,7],i_3,width=.4,color='grey',label='Out of network sms')
plt.legend(loc='upper left')
plt.xticks(range(1,8),['SUN','MON','TUE','WED','THU','FRI','SAT'])
plt.show()
```
